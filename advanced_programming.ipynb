```python
def is_matrix(A):
    size_of_rows = [len(rows) for rows in A]
    for i in range(1,len(size_of_rows)):
        if size_of_rows[i] == size_of_rows[i-1]:
            continue
        else:
            return False
    return True
A=[[1],[2,3],[4,5,]]
is_matrix(A)
```

```python
def f_x(x=1,equation = "x**2 + 3*x- 10"):
    return eval(equation)

def f_x_h(x=1,h=0.01,equation = "x**2 + 3*x- 10"):
    x = x + h
    return eval(equation)
        
def f1_derivative(x=1,h =.01,equation = "x**2 + 3*x- 10"):
    change =  f_x_h(x=x,h=h,equation = equation) - f_x(x=x, equation =equation) 
    rate_of_change = change / h
    return rate_of_change
f1_derivative(x=2,h =.01,equation = "x**2 - 10")

def newton_roots(x=0.1,tol=0.001, equation = "x**2 - 10"):
    steps = 0
    while abs(f_x(x,equation)) > tol :
        steps = steps + 1
        rate_of_change = f1_derivative(x=x,h =.01,equation = equation)
        f_x_value = f_x_h(x=x,h=0.01,equation = equation)
        x = x - f_x_value/rate_of_change
        print(x,rate_of_change,f_x_value, f_x_value/rate_of_change,sep="#")
    return (steps,x)
newton_roots(x=0.1,tol=0.001, equation = "x**2 - 10")  
```
